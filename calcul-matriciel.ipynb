## Calcul Matriciel

**Sujet : Analyse de températures.**

Imagine que tu as relevé les températures de 3 villes sur 7 jours (une matrice 3x7).

**Objectifs :**

1. **Inspection** : Affiche la forme (`shape`) et le type de données de la matrice.
2. **Écrêtage** : Remplacer toutes les valeurs inférieures à 20 par la valeur 20 
3. **Moyennes** :
    * Calcule la température moyenne globale.
    * Calcule la moyenne de chaque ville (utilise `axis=1`).

4. **Seuil** : Identifie tous les jours où la température a dépassé 30 degrés (indexation booléenne).
5. **Normalisation** : Convertis toutes les températures en degrés Fahrenheit () de manière vectorisée.

```python
import numpy as np

# Données : 3 villes (lignes), 7 jours (colonnes)
temperatures = np.array([
    [22, 25, 19, 30, 32, 28, 24], # Ville A
    [15, 18, 21, 20, 19, 17, 16], # Ville B
    [31, 34, 33, 29, 30, 35, 32]  # Ville C
])

temperatures
```

1. **Inspection** : Affiche la forme (`shape`) et le type de données de la matrice.

```python
print(temperatures.shape, temperatures.dtype)
```

2. **Écrêtage** : Remplacer toutes les valeurs inférieures à 20 par la valeur 20 

```python
# data = np.clip(temperatures, min=20)

temperatures[temperatures < 20] = 20

temperatures
```

3. **Moyennes** :
    * Calcule la température moyenne globale.
    * Calcule la moyenne de chaque ville (utilise `axis=1`).

```python
moyenne = temperatures.mean()

print("Temperature moyenne global :", moyenne)

moyenne_par_ville = temperatures.mean(axis=1)

print("Temperature moyenne par ville :", moyenne_par_ville)
```

4. **Seuil** : Identifie tous les jours où la température a dépassé 30 degrés (indexation booléenne).

```python
temperatures[temperatures > 30]
```

5. **Normalisation** : Convertis toutes les températures en degrés Fahrenheit () de manière vectorisée.

```python
temperatures = temperatures * 1.8 + 32

temperatures
```
